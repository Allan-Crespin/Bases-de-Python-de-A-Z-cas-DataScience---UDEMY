## Lire le fichier dans une liste


```python
f = open('unisex-names.csv','r')
```

```python
names = f.read()
```

```python
print(names)
```

```python
names_list = []
names_list = names.split('\n')
```

```python
print(names_list[0:10])
```

```python
names_data = []
```

```python
for name in names_list:
    value = name.split(',')
    names_data.append(value)

print(names_data[0:5])
```

```python
numerical_list = []
for num in names_data:
    name = num[0]
    value = float(num[1])
    new_list = [name, value]
    numerical_list.append(new_list)
    
```

```python
print(numerical_list)
```

```python
type(numerical_list[0][1])
```

```python
final_list = []
```

```python
for line in numerical_list:
    name = line[0]
    value = line[1]
    if value > 1000:
        final_list.append(name)
print(final_list[0:10])
        
```
