## Count female name in legislators with birth_year > 1950

```python
name_counts = {}

for row in legislators:
    gender = row[3]
    year = row[7]
    if gender == 'F' and year > 1950:
        name = row[1]
        if name in name_counts:
            name_counts[name] +=1
        else:
            name_counts[name] = 1
            
print(name_counts)
```

# None Object

```python
values = [5, 70 ,110]
max_value = 0
for i in values:
    if i > max_value:
        max_value = i
print(max_value)
```

```python
values = [-5, -70 ,-110]
max_value = 0
for i in values:
    if i > max_value:
        max_value = i
print(max_value)
```

```python
values = [-5, -70 ,-110]
max_value = None
for i in values:
    if max_value is None or i > max_value:
        max_value = i
print(max_value)
```

```python
# The highest value is indeed -5
```

## Most frequents female name in legislators

```python
max_value = None

for name in name_counts:
    count = name_counts[name]
    if max_value is None or count > max_value:
        max_value = count
        
print(max_value)
```

```python
# at most, since 1950, the number of female legislators
# who share the same name doesn't exceed 2.
```
